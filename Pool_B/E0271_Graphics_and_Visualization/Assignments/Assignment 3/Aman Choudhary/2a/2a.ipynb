```python
print("Build graph: Start")

graph = dict()
numVertices = 1558802
for currVertex in range(numVertices):
    graph.setdefault(currVertex, set())

for cell in dataset.cells_dict['tetra']:
    cell = cell.tolist()
    
    for currVertex in cell:        
        for adjacentVertex in cell:
            if(currVertex != adjacentVertex):
                graph[currVertex].add(adjacentVertex) 
            
print("Build graph: Complete")
```

```python
localMaximas = []
localMinimas = []
saddlePoints =[]

print("Finding points: Start")

iteration = 0
for currVertex in graph.keys():
    
    if iteration % 10000 == 0:
        print("iteration: " + str(iteration) + " localMaximas:" + str(len(localMaximas)) + " localMinimas:" + str(len(localMinimas)) + " saddlePoints:" + str(len(saddlePoints)) )
    iteration = iteration + 1
    
#     Scalar value of current vertex
    scalarValueCurrVertex = dataset.point_data["ttkOffsetScalarField"][currVertex]
#     print(str(currVertex) + " -> " + str(scalarValueCurrVertex)) 
    
#     Scalar value of adjacent vertices
    adjacentVertices = list(graph[currVertex])
    scalarValueAdjacentVertices = [
        dataset.point_data["ttkOffsetScalarField"][adjacentVertex]
        for adjacentVertex in adjacentVertices
    ]
    
#     for adjacentVertex in adjacentVertices: 
#         print(str(adjacentVertex) + " : " + str(dataset.point_data["ttkOffsetScalarField"][adjacentVertex])) 
#     print(scalarValueAdjacentVertices)
    
#     print("-------------------------------------")
        
#     if(iteration > 10):
#         break

    if(scalarValueCurrVertex >= max(scalarValueAdjacentVertices)):
        localMaximas += [[dataset.points[currVertex],                    
        dataset.point_data["ttkOffsetScalarField"][currVertex],                     
        dataset.point_data["MagDerivatives"][currVertex][0]]]
        continue

    if(scalarValueCurrVertex <= min(scalarValueAdjacentVertices)):
        localMinimas += [[dataset.points[currVertex],                     
        dataset.point_data["ttkOffsetScalarField"][currVertex],                     
        dataset.point_data["MagDerivatives"][currVertex][0]]]
        continue
        
    if(isMaxDisconnected(currVertex, 1) or isMinDisconnected(currVertex, 1)):
        saddlePoints += [[dataset.points[currVertex],                     
        dataset.point_data["ttkOffsetScalarField"][currVertex],                     
        dataset.point_data["MagDerivatives"][currVertex][0]]]
        continue

print("Finding points: Complete")
```

```python
class UnionFind:
    
    def __init__(self, n):
        assert n > 0, "n must be strictly positive"
        self.numElements = n
        self.parent = [i for i in range(n)]
#         self.setSize = [1 for i in range(n)]
    
    def find(self, i):
        '''Find the parent of an element.'''
        if self.parent[i] != i:
            self.parent[i] = self.find(self.parent[i])
        return self.parent[i]
    
    def union(self, x, y):
        '''Unite the two elements by uniting their parents.'''
        xparent = self.find(x)
        yparent = self.find(y)
        if xparent != yparent:
            self.parent[yparent]= xparent
            
    def is_connected(self, x, y):
        '''Check whether X and Y are connected, i.e. they have the same parent.'''
        if self.find(x) == self.find(y):
            return True
        else:
            return False
    
    @property
    def disjoint_set_count(self):
        '''Count the amount of disjoint sets.'''
        unique_parents = set([self.find(i) for i in range(self.n)])
        return len(unique_parents)
```

```python
def isMaxDisconnected(currVertex, maxSearchDepth):
    if(maxSearchDepth > 70):
        return True
    
    scalarValueCurrVertex = dataset.point_data["ttkOffsetScalarField"][currVertex]
    
    highScalarValueList = set()
    setIndex = dict()
    adjacentVertices = list(graph[currVertex])
    index = 0    
    for adjacentVertex in adjacentVertices:
        scalarValueAdjacentVertex = dataset.point_data["ttkOffsetScalarField"][adjacentVertex]
        
        if(scalarValueAdjacentVertex > scalarValueCurrVertex):
            highScalarValueList.add(adjacentVertex)
            setIndex.setdefault(adjacentVertex, index)
            index = index + 1
            
    unionFindDS = UnionFind(len(highScalarValueList))
    parents = dict()
    
    for currHighScalarValueVertex in highScalarValueList:
        
        queue = []
        queue.append(currHighScalarValueVertex)
        
        distance = dict()
        distance.setdefault(currHighScalarValueVertex, 0)
        
        while len(queue) > 0:
            currNode = queue[0]
            queue.pop(0)
            
            if(distance[currNode] >= maxSearchDepth):
                continue
                
            adjacentNodeList = list(graph[currNode])
            for adjacentNode in adjacentNodeList:
                scalarValueAdjacentNode = dataset.point_data["ttkOffsetScalarField"][adjacentNode]
                
                if(adjacentNode in highScalarValueList):
                    unionFindDS.union(setIndex[currHighScalarValueVertex], setIndex[adjacentNode])
                elif (not adjacentNode in distance) and (scalarValueAdjacentNode > scalarValueCurrVertex):
                    distance[adjacentNode] = distance[currNode] + 1
                    queue.append(adjacentNode)
                    
                    if not adjacentNode in parents:
                        parents.setdefault(adjacentNode, set())
                    parents[adjacentNode].add(currHighScalarValueVertex)
    
    for node in parents.keys():
        firstParent = parents[node].pop()
        for currParent in parents[node]:
            unionFindDS.union(setIndex[firstParent], setIndex[currParent])
    
    firstHighScalarValueVertex = highScalarValueList.pop()
    for currHighScalarValueVertex in highScalarValueList: 
        if unionFindDS.find(setIndex[firstHighScalarValueVertex]) != unionFindDS.find(setIndex[currHighScalarValueVertex]):
            return isMaxDisconnected(currVertex, maxSearchDepth+1)
        
    return False
```
