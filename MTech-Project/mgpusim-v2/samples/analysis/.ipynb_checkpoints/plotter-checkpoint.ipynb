```python
import os
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import time
warnings.filterwarnings("ignore")

def extend(index,text):
    return str(index) + '_' + str(text)

def parse(val):
    return round(float(val),2)
```

```python
statsDirPath = "/home/aman/Desktop/MTech-Project/mgpusim-v2/samples/analysis/stats/"
listDir = [
#     "matrixtranspose/2500/",
    "matrixmultiplication/2000_2000_2000/",
    "matrixmultiplication/2000_2000_2000/",
    "atax/1500_1500/",
    "bicg/1500_1500/",
    "spmv/12800_0.05/",
    "fft/64_4/",
    "relu/16777216/",
    "gemm/1500_1500_1500/"
]
```

```python
masterDict=[]

for i in range(8):
    statsDir = statsDirPath + listDir[i]
    data=dict()
    
    index=0
    for filename in os.listdir(statsDir):

        if "png" in filename:
            continue

        file = os.path.join(statsDir, filename)

        if os.path.isfile(file):
            statsFile=open(file,"r")

            index=(filename.split('_'))[0]
#             index=(filename.split('_'))[2] #for MT

            lines = statsFile.readlines()    
            for line in lines:
                line = line.split(' ')

                if(len(line)>1 and line[1] == "STAT"):
                    key=line[2]
                    value=line[4]
                    data[extend(index,key)]=value
#     print("--------------------------------------------------")
                    
    masterDict.append(data)

# print(masterDict)
# print(masterDict[0])
# for key,value in masterDict[0].items():
#     if(key.find("27")!=-1):
#         print(str(key) + " : " + str(value))
```

```python
def barplot(ax,xList,yList,plot_xlabel,plot_ylabel,plot_color,plot_title):
    ax.bar(xList,yList, color=plot_color, width = 0.32) 
    ax.set(xlabel=plot_xlabel, ylabel=plot_ylabel)
    ax.set_title(plot_title, size=10)
    ax.axhline(y = 1.0, color = '#CADCED', linestyle = '-', linewidth=1)
    
def clusturedPlot(df, axes, xlabel, ylabel, color, title, xtickLabel):
    df.plot(ax=axes, kind="bar", linewidth=0.5, edgecolor="black", width = 0.4, color=color)
    axes.set_xticklabels(xtickLabel, rotation = 0)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.set_title(title, size=10)
    axes.get_legend().remove()
```

```python
benchmarks=['MT','MM','ATAX','BICG','SPMV','FFT','RELU','GEMM']
tlb=['L1TLB','L2TLB']

def initDF():
    temp = pd.DataFrame(np.zeros((8, 2)))
    temp = temp.set_axis(benchmarks, axis='index')
    temp = temp.set_axis(tlb, axis='columns')
    return temp

tlbMissDF=initDF()
tlbMPKIDF=initDF()


for i in range(8):
    tlbMissDF.at[benchmarks[i],tlb[0]] = parse(masterDict[i][extend(27,"L1VTLB_Miss_rate")])
    tlbMissDF.at[benchmarks[i],tlb[1]] = parse(masterDict[i][extend(27,"L2TLB_Miss_rate")])
    
    tlbMPKIDF.at[benchmarks[i],tlb[0]] = parse(masterDict[i][extend(27,"L1VTLB_MPKI")])
    tlbMPKIDF.at[benchmarks[i],tlb[1]] = parse(masterDict[i][extend(27,"L2TLB_MPKI")])
    
# print(tlbMissDF)
# print(l1TLBMPKIDF)
```

```python
colorList0 = sns.color_palette("Blues", 2).as_hex()  
fig2,axes2=plt.subplots(nrows=1, ncols=1)

clusturedPlot(tlbMissDF, axes2, xlabel = "Benchmarks", ylabel = "Miss Rate %", color = colorList0, title = "TLB Miss Rate")     
# clusturedPlot(l1TLBMPKIDF, axes2, xlabel = "Benchmarks", ylabel = "MPKI", color = colorList0, title = "TLB MPKI") 

fig2.set_figwidth(15)
fig2.set_figheight(10)
fig2.tight_layout()

plt.show()
```

```python
cycleList=[]
indexList=[0,2,5,9,14,20,27]

for i in range(8):
    tempList=[]
    
    for j in range(7):
        key=extend(indexList[j],"Total_Cycles")
#         print(key)
        value=parse(masterDict[i][key])
        tempList.append(value)
    
    cycleList.append(tempList)

for tempList in cycleList:
    for i in range(7):
        tempList[i]=tempList[6]/tempList[i]
                        
```

```python
numConfigs = 28

indexKey = dict()
numberTLBs = 1
configIndex = 0

while numberTLBs <= 64:
    numberClusters = 1

    while numberClusters <= numberTLBs:
        # print(str(configIndex) + " : " + str(numberClusters) + " : " + str(numberClusters))
        indexKey[configIndex] = [numberTLBs,numberClusters]

        numberClusters *= 2
        configIndex += 1

    numberTLBs *= 2
    
print(indexKey)
```

```python
numTLBs=['1','2','4','8','16','32','64']
numClusters=['1','2','4','8','16','32','64']

def initDF3():
    temp = pd.DataFrame(np.zeros((7, 7)))
    temp = temp.set_axis(numTLBs, axis='index')
    temp = temp.set_axis(numClusters, axis='columns')
    return temp

l1MissAllDF=[]
l1MPKIAllDF=[]
l2MissAllDF=[]
l2MPKIAllDF=[]
cycleAllDF=[]

for i in range(8):
    l1TLBMissDF=initDF3()
    l1TLBMPKIDF=initDF3()
    l2TLBMissDF=initDF3()
    l2TLBMPKIDF=initDF3()
    cycleDF=initDF3()
    
#     print(l1TLBMissDF)
    baseCycle=parse(masterDict[i][extend(27,"Total_Cycles")])
    for j in range(numConfigs):
        row = str(indexKey[j][0])
        col = str(indexKey[j][1])
#         print(str(row) + " : " + str(col))

        l1TLBMissDF.at[row,col]     = parse(masterDict[i][extend(j,"L1VTLB_Miss_rate")])
        l2TLBMissDF.at[row,col]     = parse(masterDict[i][extend(j,"L2TLB_Miss_rate")])

        l1TLBMPKIDF.at[row,col]     = parse(masterDict[i][extend(j,"L1VTLB_MPKI")])
        l2TLBMPKIDF.at[row,col]     = parse(masterDict[i][extend(j,"L2TLB_MPKI")])
        
        cycleDF.at[row,col]     = baseCycle/parse(masterDict[i][extend(j,"Total_Cycles")])

    l1MissAllDF.append(l1TLBMissDF)
    l1MPKIAllDF.append(l1TLBMPKIDF)
    l2MissAllDF.append(l2TLBMissDF)
    l2MPKIAllDF.append(l2TLBMPKIDF)
    cycleAllDF.append(cycleDF)
    
# print(cycleAllDF)
```

```python
def clusturedPlot2(df, axes, xlabel, ylabel, color, title, xtickLabel):
    df.plot(ax=axes, kind="bar", linewidth=0.5, edgecolor="black", width = 1.0, color=color)
    axes.set_xticklabels(xtickLabel, rotation = 0)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.set_title(title, size=10)
    axes.get_legend().remove()
    axes.axhline(y = 1.0, color = '#CADCED', linestyle = '-', linewidth=1)
```
